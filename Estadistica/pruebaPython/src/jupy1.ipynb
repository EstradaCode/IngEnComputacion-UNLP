```python
import numpy as np

# Crear una matriz de números aleatorios
data = np.random.normal(loc=0, scale=1, size=1000)

# Calcular la media, desviación estándar y percentiles
mean = np.mean(data)
std_dev = np.std(data)
percentiles = np.percentile(data, [25, 50, 75])

print(f"Media: {mean}")
print(f"Desviación estándar: {std_dev}")
print(f"Percentiles: {percentiles}")
```

```python
import statsmodels.api as sm
import numpy as np
# Crear datos de ejemplo
X = np.random.rand(100)
y = 2 * X + np.random.randn(100) * 0.1  # y = 2*X + ruido

# Añadir una columna de constantes para la intersección
X = sm.add_constant(X)

# Ajustar el modelo de regresión lineal
model = sm.OLS(y, X)
results = model.fit()

# Resumen del modelo
print(results.summary())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
date_range = pd.date_range(start='1/1/2020', periods=100, freq='D')
time_series_data = pd.Series(np.random.randn(100).cumsum(), index=date_range)

# Visualizar la serie temporal
time_series_data.plot(title="Serie Temporal", figsize=(10, 6))
plt.show()

# Descomposición de una serie temporal con statsmodels
decomposition = sm.tsa.seasonal_decompose(time_series_data, period=7)
decomposition.plot()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm

# Prueba NumPy
data = np.random.normal(size=1000)
print(f"Media de los datos: {np.mean(data)}")

# Prueba Pandas
df = pd.DataFrame({
    'col1': np.random.randn(100),
    'col2': np.random.randn(100)
})
print(df.head())

# Prueba Matplotlib
plt.hist(data, bins=30)
plt.title("Histograma de datos")
plt.show()

# Prueba Seaborn
sns.boxplot(data=df)
plt.title("Boxplot con Seaborn")
plt.show()

# Prueba SciPy
stat, p = stats.shapiro(data)
print(f"Prueba de normalidad de Shapiro-Wilk: estadístico={stat}, p-valor={p}")

# Prueba Statsmodels
X = np.random.rand(100)
y = 2 * X + np.random.randn(100) * 0.1
X = sm.add_constant(X)  # Agregar constante para el modelo
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```
